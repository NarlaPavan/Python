```python
x = np.linspace(1,10,100)
plt.figure(1)

plt.subplot(2,2,1)
plt.title("Cubic")
plt.plot(x,x**3)

plt.subplot(2,2,2)
plt.title("Linear")
plt.plot(x,x)

plt.subplot(2,2,3)
plt.title("Log")
plt.plot(x,np.log(x))

plt.subplot(2,2,4)
plt.title("exponential")
plt.plot(x,np.exp(x))

plt.show()
```

```python
df = pd.merge(market,orders,how="inner",on="Ord_id")
df.head()
```

```python
df.info()
```

```python
df['Order_Date'] = pd.to_datetime(df['Order_Date'])
df.info()
```

```python
time_df = df.groupby(['Order_Date'])['Sales'].sum()
print(time_df)
print(type(time_df))
```

```python
plt.figure(figsize=(12,8))
sns.tsplot(data=time_df)
plt.show()
```

```python
df['month'] = df['Order_Date'].dt.month
df['year'] = df['Order_Date'].dt.year
df.head()
```

```python
df_time = df.groupby(['year','month'])['Sales'].mean()
df_time.head()
```

```python
plt.figure(figsize=(8,6))
sns.tsplot(df_time)
plt.xlabel("Time")
plt.ylabel("Sales")
plt.show()
```

```python
year_month = pd.pivot_table(df,values="Sales",index="year",columns="month",aggfunc="mean")
year_month.head()
```

```python
sns.heatmap(year_month,cmap="YlGnBu")
plt.show()
```
